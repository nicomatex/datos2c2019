```python
import pandas as pd
```

```python
df = pd.read_csv("ejercicio3.csv")
```

```python
df.head()
```

# Inciso A

```python
df.groupby("Codigo producto").agg({"Ventas Mensuales":"mean"}).reset_index()
```

# Inciso B

```python
df.groupby("Codigo producto").agg({"Ventas Mensuales":["mean","max","min"]}).stack()
```

# Inciso C

```python
df_c = df.groupby("Codigo producto").agg({"Ventas Mensuales":["min","max","mean"]}).reset_index()
df_c
```

```python
df_c.columns = ["Codigo Producto","Minimo Ventas Mensual","Maximo Ventas Mensual","Promedio Ventas Mensual"]
df_c
```

# Inciso D

```python
df_d = df.groupby("Codigo producto").agg({"Ventas Mensuales":["mean","max","min"]}).reset_index()
df_d
```

```python
df_d.columns = df_d.columns.get_level_values(0) + " " + df_d.columns.get_level_values(1)
df_d
```

```python
df_d.set_index("Codigo producto ").transpose()
```

```python
import pandas as pd
import matplotlib as plt
import seaborn as sbs
```

```python
df1 = pd.read_csv("ejercicio4_1.csv")
df2 = pd.read_csv("ejercicio4_2.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
promedios = df1.groupby("book_id").agg({"rating":"mean"}).reset_index()
```

```python
promedios
```

```python
df3 = pd.merge(promedios,df2, on='book_id', how='inner')
```

```python
df3
```

```python
df3.columns = ["book_id","avg_rating_1","book","avg_rating2"]
```

```python
df3
```

```python
df3["dif_porcentual"] = (abs(df3["avg_rating_1"] - df3["avg_rating2"]) / df3["avg_rating_1"]) * 100
```

```python
df3
```

```python
df_resultado = df3.loc[df3["dif_porcentual"] >= 20]
```
